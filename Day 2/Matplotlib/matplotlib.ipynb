```python
import matplotlib.pyplot as plt
```

## Data Loading

```python
import pandas as pd
```

```python
men_age = [28, 30, 32, 34, 36, 38, 40, 42, 44, 46, 48, 50, 45]
men_salary = [50000, 52000, 55000, 58000, 61000, 64000, 67000, 70000, 73000, 76000, 80000, 28000, 100000]
women_salary = [55000, 58000, 62000, 65000, 69000, 72000, 75000, 78000, 81000, 84000, 88000, 20000, 180000]
women_age = [25, 28, 30, 32, 35, 38, 39, 41, 44, 45, 46, 48, 43]
group = ["G1", "G2","G3","G4","G5","G6","G7","G8","G9","G10","G11","G12", "G13"]

data = pd.DataFrame({"men_age":men_age, "men_salary":men_salary,
"women_age":women_age, "women_salary":women_salary, "group":group})
data
```

## Basic Plotting

Line Plots

```python
plt.plot(data.men_age, data.men_salary, color = 'g')
plt.xlabel("Age", color = 'b')
plt.ylabel("Salary", color = 'r')
plt.title("Men's age vs Men's salary")
```

```python
plt.plot(data.men_age, data.men_salary, color = 'g', label = 'men', marker = 'o') # label is required for legend
plt.plot(data.women_age, data.women_salary, color = 'b', ls = '--', label = 'women', marker = 'x')
plt.xlabel("Age", color = 'r')
plt.ylabel("Salary", color = 'r')
plt.title("age vs salary comparision between men and women")
plt.legend()
```

Subplots

```python
plt.figure(figsize=(6, 6))
plt.subplot(1,2,1) # nrows = 2, ncols = 2, position = 1 ; (2,1,1)
plt.plot(data.men_age, data.men_salary, label="men", color="b")
plt.title("salary by age")
plt.ylabel("salary")
plt.xlabel("age")
plt.legend(loc=0) # loc=0 determines best location for the legend
plt.subplot(1,2,2) # (2,1,2)
plt.plot(data.women_age, data.women_salary, label="women", color="r")
plt.xlabel("age")
plt.ylabel("salary")
plt.legend(loc=4) # loc=4 puts the legend to bottom right
plt.tight_layout() # to fit the subplots.
```

Bar Plot

```python
plt.bar(data.group, data.men_age, color = 'lightcoral')
plt.xlabel("age groups")
plt.ylabel("salary")
plt.title("Salary by Age Groups")
```

Pie Chart

```python
categories = ['Rent', 'Groceries', 'Entertainment', 'Transportation', 'Other']
expenses = [1200, 300, 200, 100, 250]

plt.figure(figsize=(7,7))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=0) # autopct is for value labels

plt.title('Monthly Expenses Distribution')

plt.show()
```

Histogram

```python
ages = [18, 22, 25, 28, 30, 35, 40, 42, 45, 50, 55, 60, 65, 70, 75, 80]

plt.hist(ages, color = 'g', bins = 10, edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in a Population')
```

Box plot

```python
plt.boxplot([data.women_salary, data.men_salary], labels = ["Women Salary", "Men Salary"])
plt.ylabel("Salary")
```

```python
plt.boxplot([data.women_salary, data.men_salary], vert = False, patch_artist = True, labels = ["Women", "Men"])
plt.ylabel("Salary")
```
